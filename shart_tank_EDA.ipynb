```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("shark_tank_cleaned.csv")

# Setup plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
industry = df.groupby("Industry")["Total Deal Amount"].sum().sort_values(ascending=False).head(10)

plt.figure()
sns.barplot(x=industry.values, y=industry.index, palette="viridis")
plt.title("Top 10 Funded Industries")
plt.xlabel("Total Deal Amount (in Lakhs)")
plt.ylabel("Industry")
plt.tight_layout()
plt.show()
```

```python
shark_cols = [col for col in df.columns if "Investment Amount" in col and "Guest" not in col]
shark_totals = df[shark_cols].sum().sort_values(ascending=False)

plt.figure()
sns.barplot(x=shark_totals.values, y=[s.replace(" Investment Amount", "") for s in shark_totals.index], palette="crest")
plt.title("Total Investment by Each Shark")
plt.xlabel("Investment Amount (in Lakhs)")
plt.ylabel("Shark")
plt.tight_layout()
plt.show()
```

```python
offers = df["Received Offer"].value_counts()
accepted = df["Accepted Offer"].value_counts()

plt.figure()
offers.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title("Offers Received")
plt.xticks([0, 1], ["No", "Yes"], rotation=0)
plt.ylabel("Count")
plt.tight_layout()
plt.show()

plt.figure()
accepted.plot(kind='bar', color=['lightgreen', 'salmon'])
plt.title("Offers Accepted")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
top_cities = df["Pitchers City"].value_counts().head(10)
plt.figure()
sns.barplot(x=top_cities.values, y=top_cities.index, palette="mako")
plt.title("Top Cities by Number of Deals")
plt.xlabel("Number of Deals")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
presenter_deals = df.groupby("Number of Presenters")["Total Deal Amount"].count()
plt.figure()
presenter_deals.plot(kind='bar', color="goldenrod")
plt.title("Number of Presenters vs Deal Count")
plt.xlabel("Number of Presenters")
plt.ylabel("Number of Deals")
plt.tight_layout()
plt.show()
```

```python
avg_deal_by_industry = df.groupby("Industry")["Total Deal Amount"].mean().sort_values(ascending=False).head(10)
plt.figure()
sns.barplot(x=avg_deal_by_industry.values, y=avg_deal_by_industry.index, palette="flare")
plt.title("Average Deal Amount by Industry")
plt.xlabel("Average Deal Amount")
plt.ylabel("Industry")
plt.tight_layout()
plt.show()
```

```python
investment_df = df[shark_cols]
plt.figure(figsize=(12, 8))
sns.heatmap(investment_df.corr(), annot=True, cmap="coolwarm", fmt=".1f")
plt.title("Correlation Between Sharks' Investment Amounts")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=df, x="Valuation Requested", y="Total Deal Amount", hue="Industry", alpha=0.7)
plt.title("Valuation Requested vs Total Deal Amount")
plt.xlabel("Valuation Requested (Lakhs)")
plt.ylabel("Total Deal Amount (Lakhs)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
